metodi statici: 

```python
class Bus:

    wheels = 6#metodo statico perché è condiviso da tutte le istanze della classe
    
    def __init__(self, color, model, year):
        self.color = color
        self.model = model
        self.year = year

    def description(self):
        return f"{self.year} {self.color} {self.model}"
```

```python
my_bus = Bus("yellow", "volvo", 2015)
print(my_bus.wheels)
```

CLASSI DERIVATE, OVERLOADING, OVERRIDING

```python
class Truck(Bus):  # eredita da Bus
    
    def __init__(self, color, model, year, load_capacity):
        super().__init__(color, model, year)  # chiama il costruttore di Bus
        self.load_capacity = load_capacity    # attributo specifico di Truck

    def description(self):  # override del metodo
        return f"{self.year} {self.color} {self.model} with load capacity of {self.load_capacity} kg"


my_truck = Truck("white", "volvo", 2019, 15000)
print(my_truck.description())
```

Information hiding

```python
class Person:
    def __init__(self, name, age, altezza):
        self.name = name
        self.__age = age # attributo privato, non accessibile dall'esterno della classe
        self.__altezza = altezza # attributo privato, non accessibile dall'esterno della classe


    #print(p1.__age) # This will cause an error

    #getter per accedere agli attributi privati

    def get_age(self):
        return self.__age

    def get_altezza(self):
        return self.__altezza


    #setter per modificare gli attributi privati

    def set_age(self, age):
        self.__age = age

    def set_altezza(self, altezza):
        self.__altezza = altezza
```

```python
p1 = Person("Emil", 25, 170)
print(p1.get_age())        # 25

p1.set_age(26)
print(p1.get_age()) 
```

CLASSI ASTRATTE

```python
from abc import ABC, abstractmethod
#abc è una classe che permette di creare classi astratte
#abstract base classes

class Animal(ABC):
    @property
    #property serve per non dover chiamare il metodo come una funzione, ma poterlo usare come un attributo, questo è utile anche per i getter e setter
    @abstractmethod
    def species(self):
        pass 

class Dog(Animal):
    @property
    def species(self):
        return "Canine"

# Instantiate the concrete subclass
dog = Dog()
print(dog.species)
```

Per capire meglio il funzionamento del property possiamo rifare l'esempio di prima di persona però con quello

```python
class Person:
    def __init__(self, name, age, altezza):
        self.name = name
        self.__age = age          # attributo privato
        self.__altezza = altezza  # attributo privato

    # getter tramite @property
    @property
    def age(self):
        return self.__age

    @property
    def altezza(self):
        return self.__altezza

    # setter tramite @NOME.setter
    @age.setter
    def age(self, value):
        self.__age = value

    @altezza.setter
    def altezza(self, value):
        self.__altezza = value
```

```python
p1 = Person("Emil", 25, 170)

print(p1.age)        # 25
p1.age = 26          # chiama age.setter
print(p1.age)        # 26

print(p1.altezza)    # 170
p1.altezza = 175
print(p1.altezza)    # 175
```

```python
#FACTORY PATTERN:PATTERN PER CREARE MODELLI SENZA DUPLICARE CODICE TANTE VOLTE
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor


class ModelFactory:
    @staticmethod
    def create_model(model_name):
        if model_name == "linear":
            return LinearRegression()
        elif model_name == "random_forest":
            return RandomForestRegressor()
        elif model_name == "XGBoost":
            return XGBRegressor()
        else:
            raise ValueError(f"Model {model_name} not supported")

# Uso
model = ModelFactory.create_model("random_forest")
print(model)

```

```python
#STRATEGY PATTERN:PATTERN PER DEFINIRE DIVERSE STRATEGIE DI ALLENAMENTO SENZA DUPLICARE CODICE

class TrainingStrategy:#classe astratta per definire l'interfaccia della strategia di allenamento
    def train(self, model, X, y):
        raise NotImplementedError()


class SKLearnTrainingStrategy(TrainingStrategy):#strategia per allenare modelli sklearn
    def train(self, model, X, y):
        model.fit(X, y)
        return model


class ModelTrainer:#classe che utilizza una strategia di allenamento per allenare un modello
    def __init__(self, strategy: TrainingStrategy):
        self.strategy = strategy

    def run(self, model, X, y):
        return self.strategy.train(model, X, y)


# Uso
strategy = SKLearnTrainingStrategy()
trainer = ModelTrainer(strategy)
model = LinearRegression()
#trainer.run(model, X_train, y_train)
```
